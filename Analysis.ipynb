```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the CSV file
df = pd.read_csv('hash_table_set_metrics_1k.csv')

# Convert 'index' column to numeric format
df['index'] = pd.to_numeric(df['index'], errors='coerce')

# Convert 'duration' column to numeric format (in nanoseconds)
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

# Basic statistics (Nanoseconds)
print("Basic Statistics (Nanoseconds):")
print(df['duration'].describe())

# Line plot with logarithmic scaling on y-axis
plt.figure(figsize=(12, 6))
plt.plot(df['index'], df['duration'], marker='o', linestyle='-', color='b')
plt.yscale('log')  # Apply logarithmic scale
plt.title('Insertion Time Over 1k insertions (Log Scale)')
plt.xlabel('Number of Insertions')
plt.ylabel('Insertion Time (nanoseconds, Log Scale)')
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.show()

# Histogram of insertion times (in nanoseconds) with log scaling
plt.figure(figsize=(10, 6))
sns.histplot(df['duration'], bins=30, kde=True, color='g', log_scale=True)
plt.title('Histogram of Insertion Times (Nanoseconds, Log Scale)')
plt.xlabel('Insertion Time (nanoseconds)')
plt.ylabel('Frequency')
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.show()
```
